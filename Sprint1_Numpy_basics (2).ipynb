```python
# 1. Result calculator

def calc_score(marks):
  result = ""
  if marks < 40:
    result += "Fail"
  else:
    result += "Pass"
  return result

s_marks = 30
print(calc_score(s_marks))
```

```python
import pandas as pd
import numpy as np

# Read session4_dummy_students_data.csv

df = pd.read_csv("/content/session4_dummy_students_data.csv (2).xls")
print(df)
```

```python
# Create an average column using math, science, and english.

avg = df[["math","science","english"]].to_numpy()
df["average"] = avg.mean(axis=1)
print(df.head())
```

```python
# Create a result column: Pass if average >= 40, otherwise Fail.

df["result"] = np.where(df["average"] >= 40, "Pass", "Fail")
df.head()
```

```python
# Print class average.

class_avg = df["average"].mean()
print(class_avg)
```

```python
# Print highest average and lowest average.

highest_avg = df["average"].max()
lowest_avg = df["average"].min()
print(highest_avg)
print(lowest_avg)
```

```python
# Print top student or students.

top_students = df[df["average"] == highest_avg]
print(top_students["name"])
```

```python
# Print students with attendance below 75.

attendance = df[df["attendance_percent"] < 75]
print(attendance["name"])
```

```python
# Print students who submitted fewer than 6 assignments.

assignment = df[df["assignments_submitted"] < 6]
print(assignment["name"])
```

```python
# Save final output as session4_student_report.csv

df.to_csv("session4_student_report.csv", index=False)
```

```python
# 6. Today's target -

import pandas as pd
import numpy as np

df = pd.read_csv("/content/session4_dummy_students_data.csv (2).xls")

# Find the average final_project_score.

avg_project_score = df["final_project_score"].mean()

print("Average Final_Project_Score:", avg_project_score)
```

```python
# Find students with study_hours_per_week > 8.

study_hrs = df[df["study_hours_per_week"] > 8]
print(study_hrs[["name","study_hours_per_week"]])
```

```python
# Compare average marks of Mumbai students and Pune students

mumbai = df[df["city"]=="Mumbai"]
mumbai_avg = mumbai["average"].mean()

pune = df[df["city"]=="Pune"]
pune_avg = pune["average"].mean()

print("Average marks of Mumbai students:", mumbai_avg)
print("Average marks of Pune students:", pune_avg)

if len(mumbai) == 0 or len(pune) == 0:
  print("Sorry, we cannot compare!")
elif mumbai_avg > pune_avg:
  print("Mumbai students have higher average marks.")
else:
  print("Pune students have higher average marks.")
```

```python
# Create a new column called performance_level

df["performance_level"] = df["average"].apply(lambda x: "High" if x >= 75 else "Low")
df.head()
```

```python
# Save only selected columns to a new CSV

selected = df[["student_id", "name", "attendance_percent", "average","performance_level"]]
print(selected)
```

```python
#6.
students = [
    {"Stu_name":"Ishika", "marks": 88, "attendance": 95, "lab_assignment": True},
    {"Stu_name":"Diya", "marks": 75, "attendance": 45, "lab_assignment": False},
    {"Stu_name":"Sheetal", "marks": 92, "attendance": 90, "lab_assignment": True},
    {"Stu_name":"Nupur", "marks": 58, "attendance": 60, "lab_assignment": False},
    {"Stu_name":"Radha", "marks": 99, "attendance": 91, "lab_assignment": True}
]

for student in students:
    print(student["Stu_name"])

    if student["marks"] > 70:
        print("Distinction")

    if student["attendance"] < 75:
        print("Low Attendance")

    if student["lab_assignment"] == False:
        print("Lab Assignment Not Completed")

    print()
```

```python
# This project analyzes Student performance using Python, Pandas and Matplotlib.

# Features Created:
#  Total_score
#  Eligible_students
#  Need_improvement
#  Score_gap
#  Top_batch
```

```python
import pandas as pd
import matplotlib.pyplot as plt     # or from matplotlib import pyplot as plt
```

```python
data = {
    "Student_id" : [101, 102, 103, 104, 105, 106],
    "Attendance_percentage" : [92, 67, 81, 45, 74, 88],
    "Assignment_score" : [18, 12, 15, 8, 14, 19],
    "Quiz_score" : [72, 48, 65, 30, 55, 80],
    "Lab_completed" : [True, False, True, False, True, True]
}

df = pd.DataFrame(data)                                                         # converting data into row and column format
df                                                                              # print(df) will give you PLAIN UNFORMATTED TEXT
```

```python
df.shape                         # Will give you rows X columns
```

```python
df.head(3)                      # by default, it will show first 5 rows in pandas programming
```

```python
df.info()                                         # displays a concise structural summary of a DataFrame
```

```python
df.describe()     # statistical data
```

```python
# Feature creation
# 1

df["Total_score"] = df["Assignment_score"] + df["Quiz_score"]
df
```

```python
# 2

df["Eligible_students"] = ( (df["Attendance_percentage"] >= 75) &
                            (df["Total_score"] >= 70) &
                            (df["Lab_completed"])
                        )
df
```

```python
# 3

df["Need_improvement"] = df["Total_score"].apply( lambda x: 'Yes' if x < 70 else 'No')
df
```

```python
# 4

df["Score_gap"] = 100 - df["Total_score"]
df
```

```python
# 5

df["Top_batch"] = df["Score_gap"].apply(lambda x: 'Yes' if x<= 20 else 'No')
df
```

```python
# Rule-based selection

eligible = df[df['Eligible_students'] == True]   # 'True' will give error because Eligible_students contains Boolean values (True/False)
eligible
```

```python
top_students = df[df["Total_score"] >= 90]
top_students
```

```python
# Graph 1 : Analyse student performance

x = df["Student_id"]
y = df["Total_score"]
plt.plot(x,y)

plt.xlabel("Student_id")
plt.ylabel("Total_score")
plt.title("Student performance")

plt.show()
```

```python
# Student_id 106 has the highest score (99).

# Student_id 104 has the lowest score (38).
```

```python
# Graph 2 : Eligible vs Non-Eligible

d = df["Eligible_students"]
d.value_counts().plot(kind='bar')

plt.title("Eligible vs Non-Eligible")

plt.show()
```

```python
# Only half of the students passed eligibility criteria.
```

```python
# Graph 3 : Students in Top-batch

x = df["Student_id"]
y = df["Top_batch"]
plt.bar(x,y)

plt.xlabel("Student_id")
plt.ylabel("Top_batch")
plt.title("Students in Top-batch")

plt.show()
```

```python
# Student_id : 101, 103, 106 are in top-batch.
# Top-batch students have score_gap of 20 or less.
```
